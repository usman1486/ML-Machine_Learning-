```python
df_cate=d.select_dtypes(include=['object'])
```

```python
df_cate.shape
```

```python
df_cate.isnull().sum().sum()
```

```python
per_data=df_cate.isnull().sum()/df_cate.shape[0]*100
per_data
```

```python
drop_col=per_data[per_data>10].keys()
drop_col
```

```python
df=df_cate.drop(columns=drop_col)
df
```

```python
df.isnull().sum().sum()
```

```python
missing_data_col_name=[var for var in df.columns if df[var].isnull().sum()>0]
```

```python
(missing_data_col_name)
```

```python
df['Electrical'].value_counts()
```

```python
df['Electrical'].mode()
```

```python
df['Electrical'].fillna(df['Electrical'].mode()[0])
```

```python
for var in missing_data_col_name:
     (df[var].fillna(df[var].mode()[0],inplace=True))
```

```python
df.isnull().sum().sum()
```

```python
df.shape
```

```python
#Now take a look towards data distribution 
#we cant use distplot plot due to categorical data
#we histplot to see data distibution
#we cant draw histplot if there is "NAN" Value present in the dataset
plt.figure(figsize=(16,9))
for i,var in enumerate(missing_data_col_name):
     plt.subplot(4,3,i+1)
     plt.hist(df[var],label="impute",color="yellow")
     plt.hist(df_cate[var].dropna(),label="Orignal",color="grey")
     plt.legend()
        
     
     
```
