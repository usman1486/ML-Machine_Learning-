```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
d=load_breast_cancer()
```

```python
d
```

```python
data=pd.DataFrame(np.c_[d['data'],d["target"]],columns=np.append(d['feature_names'],['target']))
```

```python
data
```

```python
sns.heatmap(data)
```

```python
avc=data.corr()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(avc,annot=True,linewidth=6,linecolor="y")

plt.show()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(avc,annot=True,linewidth=6,linecolor="y")
ax.tick_params(size=5,color="r",labelsize=10,labelcolor="k")
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
dd=load_breast_cancer()
dd
```

```python
import pandas as pd
data=pd.DataFrame(np.c_[dd['data'],dd['target']],columns=np.append(dd['feature_names'],['target']))
```

```python
data
```

```python
sns.pairplot(data,vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area'],hue="target")
```

```python
sns.pairplot(data,x_vars=['mean radius', 'mean texture', 'mean perimeter', 'm],y_vars=['mean area'])

```
