```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7,8,9]
temp=[35,67,76,53,44,60,49,33,39]
```

```python
plt.plot(days,temp)

plt.xlabel("Days")
plt.ylabel("temprature")
plt.show()
```

```python
plt.plot(days,temp)
plt.axis([0,10,0,100])
plt.xlabel("Days")
plt.ylabel("temprature")
plt.show()
```

```python
plt.plot(days,temp,color="yellow",marker="o",label="TempLine")
plt.axis([0,10,0,100])
plt.xlabel("Days")
plt.ylabel("temprature")
plt.legend(loc=4)
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.title("Temprature Graph Lahore Month June ")
plt.plot(days,temp,color="yellow",marker="o",label="TempLine")
plt.axis([0,10,0,100])
plt.xlabel("Days")
plt.ylabel("temprature")
plt.legend(loc=4)
plt.grid(color="black")
plt.show()
```

```python
days=[1,2,3,4,5,6,7,8,9]
temp_lhr=[35,67,76,53,44,60,49,33,39]
temp_krc=[50,55,45,37,67,34,48,39,44]
from matplotlib import style
style.use("ggplot")
plt.title("Temprature Graph Lahore & Karachi Month June ")
plt.plot(days,temp_lhr,color="yellow",marker="o",label="Temp_Lahore")
plt.plot(days,temp_krc,color="red",marker="o",label="Temp_Karachi")
plt.axis([0,10,0,100])
plt.xlabel("Days")
plt.ylabel("temprature")
plt.legend(loc=4)
plt.grid(color="black")
plt.show()
```

```python
#histogram No need x and y axis only one axis is enough to get insights
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
ML_students=np.random.randint(18,45,(100))
PY_students=np.random.randint(15,37,(100))
```

```python
plt.hist(ML_students)
plt.show()
```

```python
#bin parameter
bins=[10,15,20,25,30,35,40,45,50]
ML_students=np.random.randint(18,45,(100))
plt.hist(ML_students,bins,rwidth=0.9)
plt.show()
```

```python
#bin parameter
bins=[10,15,20,25,30,35,40,45,50]
ML_students=np.random.randint(18,45,(100))
plt.hist(ML_students,bins,rwidth=0.9,histtype="step")
plt.show()
```

```python
#bin parameter
bins=[10,15,20,25,30,35,40,45,50]
ML_students=np.random.randint(18,45,(100))
plt.hist(ML_students,bins,rwidth=0.9,color="red",orientation="vertical",label="ML")
plt.legend(loc=2)
plt.show()
```

```python
bins=[10,15,20,25,30,35,40,45,50]
ML_students=np.random.randint(18,45,(100))
plt.hist(ML_students,bins,rwidth=0.9,color="red",orientation="vertical",label="ML")
plt.hist(PY_students,bins,rwidth=0.9,color="blue",orientation="vertical",label="PY")
plt.legend(loc=2)
plt.show()
```

```python
from matplotlib import style
bins=[10,15,20,25,30,35,40,45,50]
ML_students=np.random.randint(18,45,(100))
plt.figure(figsize=(16,9))
plt.hist([ML_students,PY_students],bins,rwidth=0.9,color=["red","blue"],orientation="vertical",label=["ML","PY"])
style.use("ggplot")
#plt.hist(PY_students,bins,rwidth=0.9,color="blue",orientation="vertical",label="PY")
plt.legend(loc=2)
plt.grid(color="black",linestyle="--")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
cls=["AI","Py","ML","DS","POP"]
grp1=[12,15,54,34,56]
grp2=[16,19,28,49,33]
grp3=[20,52,3,44,55]
```

```python
plt.bar(cls,grp1)
```

```python
plt.bar(cls,grp2)
```

```python
plt.bar(cls,grp1,color="y",width=0.4,edgecolor="k",linewidth=2,linestyle='--',alpha=0.9)
plt.bar(cls,grp2,color="k",width=0.2,edgecolor="k",linewidth=2,linestyle='--',alpha=0.9)
plt.bar(cls,grp3,color="b",width=0.3,edgecolor="k",linewidth=2,linestyle='--',alpha=0.9)
plt.show()
```

```python
plt.figure(figsize=(16,9))
cls_index=np.arange(len(cls))
width=0.2
plt.bar(cls_index,grp1,color="y",width=0.2)
plt.bar(cls_index+width,grp2,color="k",width=0.2)
plt.bar(cls_index+width+width,grp3,color="b",width=0.2)
plt.show()
```

```python
plt.figure(figsize=(16,9))
cls_index=np.arange(len(cls))
plt.bar(cls_index,grp1,color="y",width=0.2)
plt.bar(cls_index+width,grp2,color="k",width=0.2)
plt.bar(cls_index+width+width,grp3,color="b",width=0.2)
plt.xticks(cls_index,cls)
plt.show()
```

```python
#pie chart we need one axis values with the help of that we can draw that
cls=["AI","Py","ML","DS","POP"]
grp1=[12,15,54,34,56]
```

```python
plt.pie(grp1,labels=cls)
plt.show()
```

```python
a=[0,0,0.2,0,0]
plt.pie(grp1,labels=cls,explode=a)
plt.show()
```

```python
a=[0,0,0.2,0,0]
c=['g','b','r','y','k']
plt.pie(grp1,labels=cls,explode=a,colors=c)
plt.show()
```

```python
a=[0,0,0.2,0,0]
c=['g','b','r','y','k']
plt.pie(grp1,labels=cls,explode=a,colors=c,autopct="%0.0f%%")
plt.show()
```

```python
a=[0,0,0.2,0,0]
c=['g','b','r','y','k']
plt.pie(grp1,labels=cls,explode=a,colors=c,shadow=True,radius=1.5,startangle=270)
plt.show()
```
