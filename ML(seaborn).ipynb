```python
abc=np.linspace(1,100,12).reshape(4,3)
```

```python

c={"orientation":"horizontal","shrink":1,"extend":"min","extendfrac":0.1,"ticks":np.arange(0,30),"drawedges":True}
sns.heatmap(abc,linewidth=0.2,linecolor="y",annot=True,vmin=0,vmax=30,cbar_kws=c)
sns.set(font_scale=0.2)
```

```python
#coreltion heatmap used for 
#feature selection and feature engineering
#to see which feature is strongly corealated with our target variable
```

```python
#coralation
#value between-->-1...+1
#statistical measure of relationship between two variable(x,y)
#"-1"-->perfect negative corelation (x decreases then  y decreases)
#"0" no coralation
#"1":perfect positive corelatoion (direct relation)
```

```python
data=sns.load_dataset('titanic')
abc=data.corr()
plt.figure(figsize=(16,9))
ax=sns.heatmap(abc,annot=True,cbar_kws={'orientation':'horizontal','drawedges':True,'extend':"min",'shrink':1})
ax.tick_params(labelsize=20,labelcolor='b',)
plt.show()
```
