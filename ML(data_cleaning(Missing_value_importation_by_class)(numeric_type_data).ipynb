```python
num_vars=['MasVnrArea', 'GarageYrBlt']
cat_vars=["LotConfig"]
for cat_var,num_var in zip(cat_vars,num_vars):
    for var in d[cat_var].unique():
      df_num.update(d[d.loc[:,cat_var]==var][num_var].replace(np.nan,d[d.loc[:,cat_var]==var][num_var].mean()))
```
